```python

import numpy as np
import matplotlib.pyplot as plt
rho1=300
rho2=500
r1=50
z1=200
r2=200
z2=1000
V1=4*np.pi*r1**3/3
V2=4*np.pi*r2**3/3
```

```python
L=5000
dL=20
s1=1500
s2=2500
```

```python
G=6.673*1e-11
dx=np.arange(0,L,dL)
x1=dx-s1
x2=dx-s2
m1=x1**2+z1**2
n1=m1**1.5
g1=G*rho1*V1*z1/n1*1e5
m2=x2**2+z2**2
n2=m2**1.5
g2=G*rho2*V2*z2/n2*1e5
g=g1+g2
```

```python
import matplotlib.pyplot as plt
plt.plot(dx,g1,'-')
plt.plot(dx,g2,'*')
plt.title('Gravity anomaly of buried sphere model')
plt.xlabel('Distance(m)')
plt.ylabel('Gravity anomaly(mGal)')
plt.grid()
plt.show()
```

```python
p=np.arange(0,360,1)
q=np.array(p*np.pi/180)
fig,(ax1,ax2)=plt.subplots(2,1,sharex=True)
ax1.plot(dx,g)
ax1.grid()
ax1.set_ylabel("Gravity anomaly(mgal)")
ax1.set_title("Gravity anomaly of buried sphere model")
ax2.fill(r1*np.cos(q)+s1,r1*np.sin(q)+z1,"-")
ax2.fill(r2*np.cos(q)+s2,r2*np.sin(q)+z2,"-")
ax2.axis('equal')
ax2.invert_yaxis()
ax2.grid()
ax2.set_xlabel("Distance(m)")
ax2.set_ylabel("Depth(m)")
plt.show()
```
